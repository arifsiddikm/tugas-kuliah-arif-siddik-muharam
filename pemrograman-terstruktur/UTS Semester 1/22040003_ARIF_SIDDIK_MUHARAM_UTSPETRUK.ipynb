```python
# Tugas 1: Menjumlahkan Dua Angka

# Inisialisasi Input Variable AngkaPertama
AngkaPertama = int(input("Masukkan Angka Pertama : "))
# Inisialisasi Variable AngkaKedua
AngkaKedua = int(input("Masukkan Angka Pertama : "))
# Proses Menghitung
HasilJumlah = AngkaPertama + AngkaKedua
# Cetak Hasilnya
print("Hasil Penjumlahannya adalah :",HasilJumlah)

# Hasil Program
  # Masukkan Angka Pertama : 10
  # Masukkan Angka Pertama : 10
  # Hasil Penjumlahannya adalah : 20
# End 
```

```python
# Tugas 2: Menghitung Akar Kuadrat

# Inisialisasi Variable AngkaPertama
AngkaPertama = float(input("Masukkan Angka Pertama : "))
# Inisialisasi Variable AngkaKedua
AngkaKedua = 0.5
# Proses Menghitung
HasilJumlah = AngkaPertama ** AngkaKedua
# Inisialisasi Hasil
PesanHasil = "Hasilnya adalah %0.3f"
# Cetak Hasilnya
print(PesanHasil %(HasilJumlah))

# Hasil Program
  # Masukkan Angka Pertama : 10
  # Hasilnya adalah 3.162
# End 
```

```python
# Tugas 3: Menghitung Luas Segitiga

# Inisialisasi Variable AngkaPertama
Alas = float(input("Masukkan Jumlah Alas : "))
# Inisialisasi Variable AngkaKedua
Tinggi = float(input("Masukkan Jumlah Tinggi : "))
# Proses Menghitung
HasilJumlah = 0.5 * Alas * Tinggi
# Inisialisasi Hasil
PesanHasil = "Hasil dari Luas Segitiga adalah %0.2f"
# Cetak Hasilnya
print(PesanHasil %(HasilJumlah))

# Hasil Program
  # Masukkan Jumlah Alas : 15
  # Masukkan Jumlah Tinggi : 35
  # Hasil dari Luas Segitiga adalah 262.50
# End 
```

```python
# Tugas 4: Menghitung Volume Kubus

# Inisialisasi Variable AngkaPertama
Sisi_1 = int(input("Masukkan Jumlah Sisi Pertama : "))
# Inisialisasi Variable AngkaKedua
Sisi_2 = int(input("Masukkan Jumlah Sisi Kedua : "))
# Inisialisasi Variable AngkaKetiga
Sisi_3 = int(input("Masukkan Jumlah Sisi Ketiga : "))
# Proses Menghitung
HasilJumlah = Sisi_1 * Sisi_2 * Sisi_3
# Inisialisasi Hasil
PesanHasil = "Hasil dari Volume Kubus adalah %0.2f"
# Cetak Hasilnya
print(PesanHasil %(HasilJumlah))

# Hasil Program
  # Masukkan Jumlah Sisi Pertama : 10
  # Masukkan Jumlah Sisi Kedua : 10
  # Masukkan Jumlah Sisi Ketiga : 10
  # Hasil dari Volume Kubus adalah 1000.00
# End 
```

```python
# Tugas 5: Menghitung Luas Lingkaran

# Inisialisasi Variable Jari Jari Lingkaran
JariJariLingkaran = float(input("Masukkan Jumlah Jari Jari Lingkaran : "))
# Proses Menghitung
HasilJumlah = 3.14 * JariJariLingkaran * JariJariLingkaran
# Inisialisasi Hasil
PesanHasil = "Hasil dari Luas Lingkaran adalah :"
# Cetak Hasilnya
print(PesanHasil,HasilJumlah)

# Hasil Program
  # Masukkan Jumlah Jari Jari Lingkaran : 10
  # Hasil dari Luas Lingkaran adalah : 314.0
# End 
```

```python
# Tugas 6: Mengubah Kilometer jadi Mil

# Inisialisasi Variable KiloMeter
KiloMeter = float(input("Masukkan Jumlah Kilo Meter : "))
# Proses Menghitung || 1 KM = 0,62137 Mil
HasilJumlah = 0.62137 * KiloMeter
# Inisialisasi Hasil
PesanHasil = "Hasil dari Konversi KiloMeter adalah : %0.2f"
# Cetak Hasilnya
print(PesanHasil %(HasilJumlah), "Mil")

# Hasil Program
  # Masukkan Jumlah Kilo Meter : 10
  # Hasil dari Konversi KiloMeter adalah : 6.21 Mil
# End 
```
